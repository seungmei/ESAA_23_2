### Introduction:

This exercise was created based on the tutorial and documentation from [Seaborn](https://stanford.edu/~mwaskom/software/seaborn/index.html)  
The dataset being used is tips from Seaborn.

### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

```python
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv', sep=',')
data
```

### Step 3. Assign it to a variable called tips

```python
tips = data
```

### Step 4. Delete the Unnamed 0 column

```python
tips_1 = tips.drop([0])
tips_1
```

### Step 5. Plot the total_bill column histogram

```python
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.hist(tips_1['total_bill'])
axes.set_title('Histogram of Total Bill')
axes.set_xlabel('Total Bill')
axes.set_ylabel('Frequency')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
fig1 = plt.figure()
axes1 = fig1.add_subplot(1, 1, 1)

axes1.scatter(tips_1['total_bill'], tips_1['tip'])
axes1.set_title('Histogram of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig2 = plt.figure()
axes2 = fig2.add_subplot(1, 1, 1)

axes2.scatter(tips_1['total_bill'], tips_1['tip'], c = tips_1['size'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### Step 8. Present the relationship between days and total_bill value

```python
fig3 = plt.figure()
axes3 = fig3.add_subplot(1, 1, 1)

axes3.scatter(tips_1['day'], tips_1['total_bill'] , s = 0.5)
axes3.set_title('Scatterplot of Total Bill references to day ')
axes3.set_xlabel('Day')
axes3.set_ylabel('Total Bill')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
fig4 = plt.figure()
axes4 = fig4.add_subplot(1, 1, 1)

def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1

tips_1['sex_color'] = tips_1['sex'].apply(recode_sex)

axes4.scatter(tips_1['tip'], tips_1['day'] , s = 0.5, c = tips_1['sex_color'])
axes3.set_title('Scatterplot of Tip vs Day differ by Sex')
axes3.set_xlabel('Tip')
axes3.set_ylabel('Day')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips_1)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig5 = plt.figure()

tips_2  = tips_1[tips_1['time'] == 'Lunch']
tips_3 = tips_1[tips_1['time'] == 'Dinner']

axes5 = fig5.add_subplot(1, 2, 1)
axes5.hist(tips_2['tip'])
axes5.set_title('Histogram of Tip on Lunch')
axes5.set_xlabel('Tip')

axes6 = fig5.add_subplot(1, 2, 2)
axes6.hist(tips_3['tip'])
axes6.set_title('Histogram of Tip on Dinner')
axes6.set_xlabel('Tip')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig6 = plt.figure()

tips_4  = tips_1[tips_1['sex'] == 'Female']
tips_5 = tips_1[tips_1['sex'] == 'Male']

def recode_smoker(smoker):
  if smoker == 'Yes':
    return 0
  else:
    return 1

tips_4['smoker_color'] = tips_1['smoker'].apply(recode_smoker)
tips_5['smoker_color'] = tips_1['smoker'].apply(recode_smoker)

axes7 = fig6.add_subplot(1, 2, 1)
axes7.scatter(tips_4['total_bill'], tips_4['tip'], c=tips_4['smoker_color'])
axes7.set_title('Scatterplot of Total Bill VS Tip of Female differing by smoking')
axes7.set_xlabel('Total Bill')
axes7.set_ylabel('Tip')

axes8 = fig6.add_subplot(1, 2, 2)
axes8.scatter(tips_5['total_bill'], tips_5['tip'], c=tips_5['smoker_color'])
axes8.set_title('Scatterplot of Total Bill VS Tip of Male differing by smoking')
axes8.set_xlabel('Total Bill')
axes8.set_ylabel('Tip')
```
